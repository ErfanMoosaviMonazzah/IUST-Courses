```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open('./results/ipy_progresses/cnn.ipynb', 'r', encoding='utf-8') as file:
    jres = json.load(file)
```

```python
train_loss = []
train_acc = []
val_loss = []
val_acc = []
for cell in jres['cells'][0]['outputs']:
    text = cell['text'][0] 
    if 'Epoch' not in text:
        continue
    words = text.split(' ')
    if words[3]=='training':
        train_loss.append(float(words[5].replace(',','')))
        train_acc.append(float(words[-1].replace('\n','')))
    if words[3]=='validation':
        val_loss.append(float(words[5].replace(',','')))
        val_acc.append(float(words[-1].replace('\n','')))
```

```python
train_x = np.arange(120)
val_x = np.arange(1, 25) * 5 - 1
```

```python
plt.plot(train_x, train_loss)
plt.plot(val_x, val_loss, 'd-', ms=10)
```

```python
plt.plot(train_x, train_acc)
plt.plot(val_x, val_acc, 'd-', ms=10)
```
