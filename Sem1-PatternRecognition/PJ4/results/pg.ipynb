```python
import os
import pandas as pd

import matplotlib.pyplot as plt
```

```python
results = os.listdir('.')
results.remove('pg.ipynb')
results
```

```python
results
```

```python
for i in range(0, len(results), 2):
    test = pd.read_csv(results[i])
    train = pd.read_csv(results[i+1])
    
    plt.plot(train.iloc[:, 1])
    plt.plot(range(0, len(train), 200), test.iloc[:, 1])
    plt.title(results[i].split('__')[0].replace('-', ' ').replace('COS', 'COSINE Scheduler'))
    plt.savefig(f'{results[i].split("__")[0]}.png')
    plt.show()
```
